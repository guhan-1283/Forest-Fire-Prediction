## The dataset contains 2 regions 
## 0  to 122 - Bejaiya region
## next 122 - Siddi abbes region

```python
df.loc[:122,'Region'] = 0
df.loc[122:,'Region'] = 1
```

```python
df.head()
```

```python
df['Region'].value_counts()
```

```python
df.info()
```

```python
# change the region column from float to int
df['Region'] = df['Region'].astype(np.int64)
```

```python
df.info()
```

```python
# drop the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.loc[[122]]
```

```python
# drop the 122 index
df = df.drop(122).reset_index(drop=True)
```

```python
df.loc[[122]]
```

```python
# fix the spaces of columns
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
# changing the datatype
df.head()
```

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(np.int64)
```

```python
df.info()
```

```python
# change the features to float. Except the "classes" column
columns = [feature for feature in df.columns if df[feature].dtypes == "O"]

for column in columns:
    if column!='Classes':
        df[column] = df[column].astype(np.float64)
```

```python
df.info()
```

```python
# save the cleaned dataset to local disk
df.to_csv("Datasets/Algerian_forest_fires_dataset_cleaned.csv",index = False)
```

## Exploratory Data Analysis

```python
# drop the day,month and year column because it is not that much important in prediction
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# Encode the Classes category
df['Classes'].value_counts()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df['Classes'].value_counts()
```

```python
df
```

```python
# plotting the histogram
df.hist(bins=50,figsize=(10,10))
plt.show()
```

```python
# pie chart
percentages = df['Classes'].value_counts(normalize=True)*100
```

```python
class_labels = ['Fire','Not Fire']
plt.figure(figsize=(10,10))
plt.pie(percentages,labels=class_labels,autopct="%1.1f%%")
plt.title("Pie classes for Classes")
plt.legend(class_labels)
plt.show()
```

```python
# heatmap
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
